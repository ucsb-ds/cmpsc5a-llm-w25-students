```python
#Get rid of any rows with NaN values, unless only subscribers_for_last_30_days is NaN

def count_nan(row):
    count = 0
    for value in row:
        #NaN values
        if value!=value:
            count=count+1
        if value=='nan':
            count=count+1
    return count

youtube_data = youtube_data.with_column('all_nan_count', youtube_data.apply(count_nan))
```

```python
youtube_data
```

```python
def count_column_nan(value):
    if value!=value or value=='nan':
        return 1
    else:
        return 0
youtube_data = youtube_data.with_column('col_nan_count', youtube_data.apply(count_column_nan,'subscribers_for_last_30_days'))
```

```python
youtube_data
```

```python
#Get rid of any rows with NaN values, unless only subscribers_for_last_30_days is NaN

youtube_data_clean  = youtube_data.where('all_nan_count',are.equal_to,'col_nan_count')
```

```python
%%ai openai-chat:gpt-3.5-turbo

I have a table called youtube_data_clean with youtube channel statistics. I have two columns, called, highest_yearly_earnings and 
lowest_yearly_earnings. I want to create a new column called average_yearly_earnings which is the average of these two values.
Give me the solution using the datascience library in python.
```

```python
def calculate_average(x, y):
    return (x + y) / 2
youtube_data_clean = youtube_data_clean.with_column(
    'average_yearly_earnings',
    youtube_data_clean.apply(calculate_average, 'highest_yearly_earnings', 'lowest_yearly_earnings')
)
```
